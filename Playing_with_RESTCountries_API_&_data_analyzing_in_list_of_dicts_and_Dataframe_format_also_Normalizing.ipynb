```python


import requests
import pandas as pd
import datetime
```

```python

# Get the current date
today = datetime.date.today().strftime('%d-%m-%Y')

# Set the API endpoint and parameters
url = "https://restcountries.com/v3.1/independent"

# Send a GET request to the API endpoint
response = requests.get(url)

```

```python
# prompt: how to check type of response object
print(type(response))
print(type(response.json()))
```

```python
df = pd.DataFrame(response.json())
df.head(1)
```

```python
# prompt: drop tld, cca2, ccn3, cca3, cioc columns and display 1 member of head()

df.drop(columns=['tld', 'cca2', 'ccn3', 'cca3', 'cioc'], inplace=True)
df.head(1)
```

```python
# prompt: drop all columns except name, unMember and Capital and languages

df.drop(columns=df.columns.difference(['name', 'unMember', 'capital']), inplace=True)
df.head(1)
```

```python
# prompt: convert this dataframe to list of dictionaries
print(df)
df_dict = df.to_dict('records')
print(df_dict)
```

```python
# prompt: normalize the result dict to a df called df2 and print it

# Normalize the result dict to a DataFrame
df2 = pd.json_normalize(df_dict)

# Print the DataFrame
print(df2.columns)
print(df2.head())
```
